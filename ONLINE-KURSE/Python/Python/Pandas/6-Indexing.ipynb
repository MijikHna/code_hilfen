## 5 - Using loc[]
* `loc[]` - Label-basierter Index. wirft KeyError, wenn Item nicht gefunden
* eigentlich nach Array Index auswÃ¤hlen (per default ist ja nummerischer Index und einmalig, wenn man aber Index wechselt mit `set_index` kann dann Index symbolisch und nicht mehr einmalig sein)

```python
oo.loc['BOLT, Usain'] # Fehler, da Index der Default index ist (also nummerischer + einmaliger)
```

```python
# loc[] Zugriff
oo.set_index('Athlete', inplace=True)
oo.loc['BOLT, Usain']
```

```python
# traditioneller Zugriff
oo.reset_index(inplace=True)
oo[oo.Athlete == 'BOLT, Usain']
```

```python
oo.loc[oo.Athlete == 'BOLT, Usain']
```

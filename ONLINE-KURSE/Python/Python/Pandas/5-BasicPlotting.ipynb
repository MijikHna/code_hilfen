## 2 - Plot types
* `plot()` - plottet
    * `plot(kind='line')` - line ist auch default
    * `plot(kind='bar')` - um Gruppen zu vergleichen
    * `plot(kind='barh')` - horisontale Bars
    * `plot(kind='pie')` 

```python
import pandas as pd
import pandas

import matplotlib.pyplot as plt
%matplotlib inline

oo = pd.read_csv('./ALL MEDALISTS.csv', skiprows=4)
```

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

```python
fo.Sport.value_counts()
```

```python
fo.Sport.value_counts().plot();
```

```python
fo.Sport.value_counts().plot(kind='line');
```

```python
fo.Sport.value_counts().plot(kind='bar');
```

```python
fo.Sport.value_counts().plot(kind='barh');
```

```python
fo.Sport.value_counts().plot(kind='pie');
```

## 4 - Figsize

## 5 - Colormaps
* Colormaps sind in Matplotlib integriert
* Drei Klassen  
    * Sequential = Gradient => wenn Daten sich langsam √§ndern
    * Diverging = wenn Daten um Mittelpunkt liegen
    * Qualitative = Info, die keine Beziehungen zu einander haben
* `colormap='...'`
* man muss schauen, ob `kind` ein `color` oder `colormap` braucht
* => Doku gucken, welche Namen die Colormaps der drei Klassen haben

```python
fo.Sport.value_counts().plot(kind='pie', colormap='Pastel1');
```

## 6 - Seaborn basic plotting
* = Lib basierend auf Matplotlib
* bessere statistische Plots
* ~ Erweiterung von Matplotlib
* auf deren Seite kann man Bsp f√ºr plots ansehen
* Seaborn anstelle von Matplotlib benutzen, wenn man kompliziertere Plots machen will

```python
import seaborn as sns
```

### `countplot`
* hat √§hnliche parameter wie matplotplotlib
* `seaboard.countplot(x=None, y=None, hue=None, data=None, order=None, order=None, hue_order=None, orient=None, color=None, palette=None)`
    * `data` = DataFrame zum Ploten
    * `hue` - Kategorien f√ºr den Plot => Kategorie bekommt eigene Farbe, dann Stats der Kategorie haben unterschiedlen Gradien
    + `order` - Sequence wenn man hue benutzt
    * `palette` = Farben

```python
sns.countplot(x='Medal', data=oo);
```

```python
sns.countplot(x='Medal', data=oo, hue='Gender');
# wie man sieht x wird noch in men udn women utnerteilt
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

oo: pd.core.frame.DataFrame = pd.read_csv('ALL MEDALISTS.csv', skiprows=4)
oo.head()
```

```python
# 1.1
team_china: pd.core.frame.DataFrame = oo[(oo.Edition == 2008) & (oo.NOC == 'CHN')]
team_china.Gender.value_counts()
```

```python
team_china.Gender.value_counts().plot(kind='bar');
```

```python
#1.2
sns.countplot(data=team_china, x='Gender');
```

```python
# 2
sns.countplot(data=oo, x='Gender', palette='bwr');
```

```python
sns.countplot(x='Medal', data=team_china, hue='Gender');
```

```python
sns.countplot(x='Medal', data=team_china, hue='Gender', palette='bwr', order=['Gold', 'Silver', 'Bronze']);
```
